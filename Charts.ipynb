# Create chart of global average temperatures with data given by Nasa

Link database: https://go.nasa.gov/3LNaiA7

```python
# Read database
df = pd.read_csv(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Data\GLB.Ts+dSST.csv", header=1)
df
```

```python
# Drop the last row since it doesn't contain all the datapoints
df = df[:142]
df
```

```python
# Remove all the cols relative to months
cols = ['Year','SON']
df = df[cols]
df
```

```python
# Rename the 2nd col
df = df.rename(columns = {'SON':'Mean'})
```

```python
# Check col names
df.columns
```

```python
# Convert to float the 2nd col
df['Mean'] = df['Mean'].astype(float, errors = 'raise')
```

```python
# Convert to integer the 1st col
df['Year'] = pd.to_numeric(df['Year'])
```

```python
# Create the chart
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x='Year',y='Mean',lw=3)
plt.xlabel('Year')
plt.xlim(1880,2021)
plt.ylabel('Deviation from the average (C°)')
# plt.title('Average annual temperature anomaly',fontsize=15,y=1.05)
plt.annotate('Source: own elaborations from https://go.nasa.gov/3LNaiA7', (0,-0.1), (0,-30), fontsize=14,
             xycoords='axes fraction', textcoords='offset points')
# Save figure
plt.savefig(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Charts\Average_annual_temperature_anomaly.png",bbox_inches='tight')
```

# Create a chart representing CO2 emissions per capita for Australia, US, Canada and European Union from 1990 to 2019

```python
# Link database: https://bit.ly/world-bank-database
```

```python
# Read the database
df_3 = pd.read_csv(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Data\CO2 emissions (metric tons per capita).csv", header=2)
df_3.head()
```

```python
# See cols name
df_3.columns
```

```python
# Drop not useful cols
df_3 = df_3.drop(['Indicator Name','Indicator Code','Country Code','Unnamed: 66'], axis=1)
df_3
```

```python
# Select countries
condition = df_3['Country Name'].isin(['Australia','Canada','United States','European Union'])
df_3 = df_3[condition]
df_3
```

```python
# Start from common year
df_3 = df_3.dropna(axis=1)
df_3
```

```python
# Shape Database
df_3 = pd.melt(df_3, id_vars='Country Name', var_name='Year', value_name='CO2 emission')
df_3
```

```python
# Set index
df_3.set_index('Country Name')
```

```python
# Check variable type
df_3.info()
```

```python
# Convrt to integer YEAR
df_3['Year'] = df_3['Year'].apply(int)
```

```python
# Chart
plt.figure(figsize=(10,6))
sns.lineplot(x='Year',y='CO2 emission',hue='Country Name',data=df_3,lw=3)
# Label
plt.xlabel('Year')
plt.xlim(1990,2019)
plt.ylabel('$CO_2$ emissions per-capita, in tons')
# plt.title('Annual $CO_2$ emissions per-capita from 1990 to 2019',fontsize=15,y=1.05)
plt.legend(bbox_to_anchor = (1.25,1))
plt.annotate('Source: own elaborations from https://bit.ly/world-bank-database', (0,-0.1), (0,-30), fontsize=14,
             xycoords='axes fraction', textcoords='offset points')
# Save figure 
plt.savefig(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Charts\co2_emissions.png",bbox_inches='tight')
```

# Create a bar plot representing the evolution of time of weather-distructive events in USA from 1980 to 2022

Link database: https://bit.ly/ncei-data

```python
# Read the database
df_4 = pd.read_csv(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Data\time-series-US-1980-2022.csv", header=2)
df_4.head()
```

```python
# Cols name
df_4.columns
```

```python
# Select columns
df_4 = df_4[['Year','All Disasters Count','All Disasters Cost']]
```

```python
df_4.head()
```

```python
# Check types
df_4.info()
```

```python
# Chart
fig, ax = plt.subplots(figsize=(10,6))

# Event count
sns.lineplot(data=df_4,x='Year',y='All Disasters Count',lw=3,color='royalblue')
ax.set_ylabel("Number of events",fontsize=14,color='royalblue')
plt.xlim(1980,2021)
plt.ylim(0, 40)

# Damages
ax2 = plt.twinx()
sns.lineplot(data=df_4,x='Year',y='All Disasters Cost',lw=3,ls='--',color='firebrick')
ax2.set_ylabel("Damages, in billion USD",fontsize=14,color='firebrick')
plt.xlim(1980,2021)
plt.ylim(0, 400)

# Label
# plt.title('Weather destructive events and damages from 1980 to 2021',fontsize=15,y=1.05)
plt.annotate('Source: own elaborations from https://bit.ly/ncei-data', (0,-0.1), (0,-30), fontsize=14,xycoords='axes fraction', textcoords='offset points')

# Save figure 
plt.savefig(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Charts\event-usa.png",bbox_inches='tight')
```

# Creating a representation of some dimensions of climate change using ESS8 topline results as a source after having applied a weight and a recoding

```python
# Chart climate dimensions
fig, axes = plt.subplots(figsize=(10,7),nrows=2,ncols=2)

# fig.suptitle('Perception about reality, cause, effects and worrying about climate change across European regions',fontsize=15,y=1.01)

sns.boxplot(ax=axes[0,0],data=ess_topline,x='region',y='clmchng',order=['North','South','Centre-West','East'])
axes[0,0].set_xlabel('Region',fontsize=15)
axes[0,0].set_ylabel('Reality',fontsize=15)

sns.boxplot(ax=axes[0,1],data=ess_topline,x='region',y='ccnthum',order=['North','South','Centre-West','East'])
axes[0,1].set_xlabel('Region',fontsize=15)
axes[0,1].set_ylabel('Human activity',fontsize=15)

sns.boxplot(ax=axes[1,0],data=ess_topline,x='region',y='ccgdbd',order=['North','South','Centre-West','East'])
axes[1,0].set_xlabel('Region',fontsize=15)
axes[1,0].set_ylabel('Impact',fontsize=15)

sns.boxplot(ax=axes[1,1],data=ess_topline,x='region',y='wrclmch',order=['North','South','Centre-West','East'])
axes[1,1].set_xlabel('Region',fontsize=15)
axes[1,1].set_ylabel('Worry',fontsize=15)

plt.tight_layout() # automatic layout

# Label
axes[1,0].annotate('Source: own elaborations from https://bit.ly/eu-energy-preferences', (0,-0.1), (0,-45), fontsize=14,xycoords='axes fraction', textcoords='offset points')

# Save figure 
plt.savefig(r"C:\Users\marco\OneDrive - Microsoft 365\Thesis Project\Charts\ess8-1.png",bbox_inches='tight')
```
